```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("C:/Users/predator/Desktop/US_Accidents_March23.csv")
data.head()
```

```python
data.shape
```

```python
data.info()
```

```python
data.drop(columns = ['ID','Description','County','Zipcode','Timezone','Airport_Code','Pressure(in)',
                     'Weather_Timestamp','Sunrise_Sunset','Wind_Chill(F)','Humidity(%)',
                     'Wind_Direction','Civil_Twilight','Nautical_Twilight','Astronomical_Twilight'], inplace=True)
```

```python
data.head()
```

```python
data.isna().sum()
```

```python
data.dropna(inplace = True)
```

```python
data.isna().sum()
```

```python
data.describe()
```

```python
accidents_by_cities = data['City'].value_counts()
accidents_by_cities
```

```python
accidents_by_cities[:10]
```

```python
fig, ax = plt.subplots(figsize=(8,3))
accidents_by_cities[:10].plot(kind='bar', color='g')
ax.set(title = 'Top 10 cities By Number of Accidents',
       xlabel = 'Cities',
       ylabel = 'Accidents Count')
plt.show()
```

```python
data['Severity'].value_counts()
```

```python
sns.countplot(x='Severity', data=data)
plt.title('Distribution of Accident Severity')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.scatterplot(x='Start_Lng', y='Start_Lat', data=data, alpha=0.1)
plt.title('Geographical Distribution of Accidents')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

```python
plt.figure(figsize=(7, 4))
sns.boxplot(x='Severity', y='Temperature(F)', data=data)
plt.title('Relationship between Temperature and Accident Severity')
plt.show()
```

```python
f,ax=plt.subplots(2,3,figsize=(10,10))
ax[0,0] = data['Amenity'].value_counts().plot.pie(autopct='%1.1f%%',ax=ax[0,0],colors=["cornflowerblue","salmon"])
ax[0,0].set_title("Amenity",fontweight ="bold")

ax[0,1] = data['Crossing'].value_counts().plot.pie(autopct='%1.1f%%',ax=ax[0,1],colors=["cornflowerblue","salmon"])
ax[0,1].set_title("Crossing",fontweight ="bold")

ax[0,2] = data['Railway'].value_counts().plot.pie(autopct='%1.1f%%',ax=ax[0,2],colors=["cornflowerblue","salmon"])
ax[0,2].set_title("Railway",fontweight ="bold")

ax[1,0] = data['Station'].value_counts().plot.pie(autopct='%1.1f%%',ax=ax[1,0],colors=["cornflowerblue","salmon"])
ax[1,0].set_title("Station",fontweight ="bold")

ax[1,1] = data['Stop'].value_counts().plot.pie(autopct='%1.1f%%',ax=ax[1,1],colors=["cornflowerblue","salmon"])
ax[1,1].set_title("Stop",fontweight ="bold")

ax[1,2] = data['Traffic_Signal'].value_counts().plot.pie(autopct='%1.1f%%',ax=ax[1,2],colors=["cornflowerblue","salmon"])
ax[1,2].set_title("Traffic_Signal",fontweight ="bold")
plt.show()
```
