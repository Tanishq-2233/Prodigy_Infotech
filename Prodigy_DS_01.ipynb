```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
ages = np.random.randint(21, 90, size=250)
```

```python
plt.figure(figsize=(7, 3))
plt.hist(ages, bins=10, color='green', edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Distribution of Ages in a population')
plt.show()
```

```python
population_size = 1000
gender_sample = np.random.choice(['Male', 'Female'], size=population_size)
gender_count = np.unique(gender_sample, return_counts=True)

plt.bar(gender_count[0], gender_count[1], color=['red', 'blue'])
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Distribution of Genders in Population')

plt.show()
```
